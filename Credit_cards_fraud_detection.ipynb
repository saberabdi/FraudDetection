```python
def Distance(X,Y):
  distance=0
  for k in range (len(X)):
    distance+=(X[k]-Y[k])**2
  return sqrt(distance)
```

```python
class imb_knn():
  def __init__(self,k=5):
    self.x=None
    self.y=None
    self.k=k

  def train(self,X,Y):
    self.x=X
    self.y=Y
  def predict(self,X,ratio):
    Output=[]
    for k in range(len(X)):
      distance=[]
      for i in range (len(self.x)):
        distance.append((i,Distance(self.x[i],X[k])))
      neighbors=sorted(distance,key=first)[:self.k]
      dict={}
      dict[0]=0
      dict[1]=0
      for l in neighbors:
        if self.y[l[0]]==0:
          dict[0]+=1
        else:
          dict[1]+=ratio
      if dict[0]> dict[1]:
        Output.append(0)
      else:
        Output.append(1)
    return Output
  def score(self,X,Y,ratio):
    Output=self.predict(X,ratio)
    right=0
    for k in range(len(Y)):
      if Output[k]==Y[k]:
        right+=1
    return right/len(Y)

        





```

```python
knn=imb_knn()
```

```python
knn.train(X_train,y_train)
```

```python
knn.score(X_test,y_test,1)
```
